```python
import pandas

data = pandas.read_csv("datasets/supermarket_data.csv")
data
```

```python
# Use the describe function to get a summary of the data
data.describe()
```

```python
# We have noticed that the column item price column contains 3 empty slots.
# We need to fill these empty slots with a value and in this case we shall fill with the mean of the column
data["itemprice"] = data["itemprice"].fillna(55)

# Describe the data again to confirm that there are no more empty slots in the item price column
data.describe()
```

```python
# Get the number of entries in a given column
data.groupby("gender").size()
```

```python
# Get the number of entries in a given column
data.groupby("customertype").size()
```

```python
# Get the number of entries in a given column
data.groupby("city").size()
```

```python
# Combining multiple columns
data.groupby(["city", "payment"]).size()
```

```python
# Import the modules
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Below is a count plot showing the distribution of males to females
sns.countplot(x="gender", data=data, palette="dark")
plt.title("Gender Distribution Against Total Count")
plt.ylabel("Total Count of Individuals")
plt.xlabel("Gender")
```

```python
# Create a count plot of product
# Which product category appears the most?
sns.countplot(x="product", data=data, palette="dark")
plt.xticks(rotation=45)
plt.title("Product Distribution Against Total Count")
plt.ylabel("Total Count of Individuals")
```

```python
# Below is a countplot that has two coloumns
sns.countplot(x="gender", hue="city", palette="Blues", data=data)
```

```python
# below is a pieplot
data["payment"].value_counts().plot.pie(autopct='%1.2f%%')
```

```python
import seaborn as sns

import matplotlib.pyplot as plt

# 1. Create the plot and assign it to 'ax'

ax = sns.countplot(x="gender", data=data, palette="spring")

# 2. Add labels to each bar container

for container in ax.containers:

    ax.bar_label(container)

# 3. Rest of your formatting

plt.title("Gender Distribution Against Total Count")

plt.ylabel("Total Count of Individuals")

plt.show()
```
